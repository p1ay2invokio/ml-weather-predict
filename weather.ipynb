```python
import pandas as pd
import sklearn as sk
import xgboost as xg
```

```python
weather_data = pd.read_csv('./seattle-weather.csv')
```

```python
weather_data
```

```python
features = weather_data[["precipitation", "temp_max", "temp_min", "wind"]]
target = weather_data["weather"].map({
    'sun': 0,
    'rain': 1,
    'fog': 2,
    "drizzle": 3,
    "snow": 4
}).astype(int)
```

```python
preprocess = sk.compose.ColumnTransformer([
    ('num', sk.impute.SimpleImputer(), ["precipitation", "temp_max", "temp_min", "wind"])
])
```

```python
model = sk.pipeline.Pipeline([
    ('preprocess', preprocess),
    ('clf', xg.XGBClassifier())
])
```

```python
x_train, x_test, y_train, y_test = sk.model_selection.train_test_split(features, target, test_size=0.2)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(pd.DataFrame([[10.9,10.6,2.8,4.5]],columns=["precipitation", "temp_max", "temp_min", "wind"]))
```
